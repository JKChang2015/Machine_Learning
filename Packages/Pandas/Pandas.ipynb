```python
# Filt some of the columns
# in this case- end up with 'g'
col_names = food_info.columns.tolist()  #get columns in the list
print(col_names)
print('\n\n')

gram_columns = []  #get conditional index

for col in col_names:
    if col.endswith('(g)'):
        gram_columns.append(col)

gram_data = food_info[gram_columns]
print(gram_data[0:10])
```

```python
# basic operations
new = pd.DataFrame()
new['Sodium_(mg)'] = food_info['Sodium_(mg)']
new['add_100'] = food_info['Sodium_(mg)'] + 100
new['sub_100'] = food_info['Sodium_(mg)'] - 100
new['multiple_2'] = food_info['Sodium_(mg)'] * 2
new['div_1000'] = food_info['Sodium_(mg)'] / 1000
print(new[0:10])
```

```python
# operations between DataFrames
new = pd.DataFrame()
new['water_energy'] = food_info['Water_(g)']
new['Energ_Kcal'] = food_info['Energ_Kcal']
new['total'] = new['water_energy'] * new['Energ_Kcal']
print(new[0:10])
```

```python
# max(), min(), mean()
# normalization
max_calories = food_info["Energ_Kcal"].max()
# min_calories = food_info["Energ_Kcal"].min()
# Divide the values in "Energ_Kcal" by the largest value.
normalized_calories = food_info["Energ_Kcal"] / max_calories
normalized_protein = food_info["Protein_(g)"] / food_info["Protein_(g)"].max()
normalized_fat = food_info["Lipid_Tot_(g)"] / food_info["Lipid_Tot_(g)"].max()
food_info["Normalized_Protein"] = normalized_protein
food_info["Normalized_Fat"] = normalized_fat
```

```python
# sort()
#By default, pandas will sort the data by the column we specify in ascending order and return a new DataFrame
# Sorts the DataFrame in-place, rather than returning a new DataFrame.
# print( food_info["Sodium_(mg)"])
food_info.sort_values("Sodium_(mg)", inplace=True)
print(food_info["Sodium_(mg)"][0:10])  # ascending， default = ture
print('------------------------')
a = food_info.sort_values("Sodium_(mg)", ascending=False)  #descending， default = ture
print(a["Sodium_(mg)"][0:10])
```
