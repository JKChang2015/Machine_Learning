```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
data = pd.read_csv('../large file/creditcard.csv')
data.head()
```

```python
data.shape
```

```python
# Class =0 normal trasaction, class = 1 abnormal trasaction
count_classes = pd.value_counts(data['Class'],sort=True).sort_index()
print(count_classes)
count_classes.plot(kind = 'bar')
plt.title("Fraud class histogram")
plt.xlabel("Class")
plt.ylabel("Frequency")
```

```python
# 下采样：对多的数据进行随机采样，变成均衡数据，同样少
# 过采样：对于少的数据生成更多的数据，变成均衡数据，同样多
```

```python
# Normalization
from sklearn.preprocessing import StandardScaler

data['normAmount'] = StandardScaler().fit_transform(data['Amount'].reshape(-1, 1))
data = data.drop(['Time','Amount'],axis=1)
data.head()
```

### Decimation

```python
# Pattern values: X
# Results: y
X = data.ix[:, data.columns != 'Class']
y = data.ix[:, data.columns == 'Class']

# Number of data points in the minority class
number_records_fraud = len(data[data.Class == 1])

#save the index of the abnormal
fraud_indices = np.array(data[data.Class == 1].index)  

# random select, randomly select "x" number (number_records_fraud)
normal_indices = data[data.Class == 0].index
random_normal_indices = np.random.choice(normal_indices, number_records_fraud, replace = False)
#save the index of the normal value
random_normal_indices = np.array(random_normal_indices)

# merge abnormal & selected normal index
under_sample_indices = np.concatenate([fraud_indices,random_normal_indices])

# extract the sub set of the data (normal:abnormal = 1:1)
under_sample_data = data.iloc[under_sample_indices,:]

X_undersample = under_sample_data.loc[:, under_sample_data.columns != 'Class']
y_undersample = under_sample_data.loc[:, under_sample_data.columns == 'Class']

# Showing ratio
print("Percentage of normal transactions: ", len(under_sample_data[under_sample_data.Class == 0])/len(under_sample_data))
print("Percentage of fraud transactions: ", len(under_sample_data[under_sample_data.Class == 1])/len(under_sample_data))
print("Total number of transactions in resampled data: ", len(under_sample_data))
```
